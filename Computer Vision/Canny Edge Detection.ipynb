```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img=cv2.imread('sammy_face.jpg')
```

```python
plt.imshow(img)
```

```python
#straight to canny edge detection without blurring and trial threshold values
edges=cv2.Canny(img,threshold1=127,threshold2=127)
```

```python
#edges of the dog are displayed correctly but alone with lot of noise(which r actlly not edges)
plt.imshow(edges)
```

```python
#need to blur the image and also better to have a formula to define the thresholds instead of having arbitrary values
#calculating the median first
med=np.median(img)
```

```python
med
```

```python
#setting the threshold values
#what percent of the median values to be selected is completely arbitary and can be varied for btr results
lower=int(max(0,0.7*med))
higher=int(min(255,1.3*med))
```

```python
edges=cv2.Canny(img,threshold1=lower,threshold2=higher)
plt.imshow(edges)
```

```python
#blurring the image prior for better edges
blur_img=cv2.blur(img,ksize=(5,5))
```

```python
edges=cv2.Canny(blur_img,threshold1=lower,threshold2=higher+60)
plt.imshow(edges)
```

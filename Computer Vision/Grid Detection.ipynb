```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
flat_chess=cv2.imread('flat_chessboard.png')
```

```python
plt.imshow(flat_chess)
```

```python
#function to find the corners
#(7,7)indicates the size of img(excluding the outer boxes)
found,corners=cv2.findChessboardCorners(flat_chess,(7,7))
```

```python
found
```

```python
cv2.drawChessboardCorners(flat_chess,(7,7),corners,found)
```

```python
plt.imshow(flat_chess)
```

```python
newimg=cv2.imread('dot_grid.png')
```

```python
plt.imshow(newimg)
```

```python
#working on the dots
found,corners=cv2.findCirclesGrid(newimg,(10,10),cv2.CALIB_CB_SYMMETRIC_GRID)
#cv2.CALIB_CB_SYMMETRIC_GRID flag that indicates the patter that has to be detected.
```

```python
cv2.drawChessboardCorners(newimg,(10,10),corners,found)
```

```python
plt.imshow(newimg)
```

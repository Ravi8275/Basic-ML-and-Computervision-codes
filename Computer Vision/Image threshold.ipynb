```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
#reading the image
img=cv2.imread('rainbow.jpg')
```

```python
plt.imshow(img)
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#turning the image into gray inorder to process it for image segmentation
plt.imshow(gray_image,cmap='gray')
```

```python
#127 will be the threshold value below it will be termed as zero and the intensity value above it will be filled with max value i.e 255
ret,threshimg=cv2.threshold(gray_image,127,255,cv2.THRESH_BINARY)
#selecting the very basic binary threshold
```

```python
plt.imshow(threshimg,cmap='gray')
```

```python
img2=cv2.imread('crossword.jpg',0)
```

```python
plt.imshow(img2,cmap='gray')
```

```python
def show_pic(img):
    fig=plt.figure(figsize=(15,15))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
show_pic(img2)
```

```python
#performing binary threshold operations on the image
re,th2=cv2.threshold(img2,160,255,cv2.THRESH_BINARY)
```

```python
show_pic(th2)
```

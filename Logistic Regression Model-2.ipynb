<span style="font-size: larger;">**Training and Testing the data**.</span>

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
X=Finaldata.drop('HeartDisease',axis=1)
y=Finaldata['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
sb.countplot(x='HeartDisease',data=finaldata)
```

<span style="font-size: larger;">**The dataset here is imbalanced as the two target classes of a target variable are not balanced and way too far in number.So weight sampling is performed here**.</span>

```python
sample_weights = [1 if label == 0 else 10 for label in y_train]
lr = LogisticRegression()
lr.fit(X_train, y_train, sample_weight=sample_weights)
```

```python
pred=lr.predict(X_test)
```

<span style="font-size: larger;">**Validating the Model**.</span>

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
confusion_matrix(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```

<span style="font-size: larger;">**The Model performed well in predicting 'NO''s but struggled in predicting 'yes'.**.</span>

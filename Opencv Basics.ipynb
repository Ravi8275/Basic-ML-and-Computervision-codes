Reason for the inappropriate colour of the image is the difference in the color readability of matplotlib and opencv

```python
#for matplotlib Red blue green
#for opencv green blue red
#however opencv has inbuilt method to rectify this
new_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(new_img)
```

if we wish to read the image directly in a gray scale color format

```python
gray_img=cv2.imread('/Users/raviirt/Downloads/Computer-Vision-with-Python/DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(gray_img,cmap='gray')
```

Resizing the image

```python
resized_img=cv2.resize(new_img,(1000,400))
```

```python
plt.imshow(resized_img)
```

Resizing the image by ratio

```python
w_ratio=0.75
h_ratio=0.75
#i.e reducing the size by 75%
```

```python
ratio_sized=cv2.resize(new_img,(0,0),new_img,w_ratio,h_ratio)
```

```python
plt.imshow(ratio_sized)
```

Flipping the image

```python
flipped_img=cv2.flip(new_img,0)
#0 as parameter indicates the vertical flip
```

```python
plt.imshow(flipped_img)
```

```python
flipped_img=cv2.flip(new_img,1)
#1 as parameter indicates the vertical flip
plt.imshow(flipped_img)
```
